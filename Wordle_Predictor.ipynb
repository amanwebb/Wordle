```python
# create super long list of the words read in
with open('words.csv', newline='') as f:
  reader = csv.reader(f)
  words = list(reader)
```

*NEED TO ADJUST FOR DUPLICATE LETTERS.*
*FOR NOW, IF YOU USE A DUPLICATE LETTER AND AT LEAST ONE IS RIGHT, PUT A YELLOW*

```python
# Create empty list
word_list = []
for word in words:
  word_list.append(word[0])

# Prompt user to enter first word with recommendation
word_guess = ''

temp_word_list = []

guesses = 0

while guesses < 6:

  # actually input the word and make lowercase
  word_guess = input('Enter your guess: ').lower()

  # input the results of the guess    
  result = input('Enter the last result: ').lower()

  # if statement for when you win    
  if result == 'ggggg':
    print('\n \nCONGRATS! You won!!')
    break
  counter = 0

  for i in result:
    # g means green aka correct letter & correct space      
    if i == 'g':
      # check words in word bank for letters matching the correct letter and 
      # space guessed and make reformed word bank with only those words
      for word in word_list:
          if word[counter] == word_guess[counter]:
            temp_word_list.append(word)
      del word_list
      word_list = temp_word_list.copy()
      del temp_word_list
      temp_word_list = []

    counter = counter + 1
  counter = 0

  for i in result:
    # b means not in word at all
    if i == 'b':
      # check words in word bank for letters matching the incorrect 
      # letter guessed and delete all words with that letter
      for word in word_list:
        if word_guess[counter] in word:
          pass
        else:
          temp_word_list.append(word)
      del word_list
      word_list = temp_word_list.copy()
      del temp_word_list
      temp_word_list = []

    counter = counter + 1
  counter = 0

  for i in result:
    # y means letter is in the word but not in correct space
    if i == 'y':
      # check words in word bank for letters matching the correct letter but 
      # not space guessed and make reformed word bank with only those words      
      for word in word_list:
        if word_guess[counter] in word and word[counter] != word_guess[counter]:
          temp_word_list.append(word)
      del word_list
      word_list = temp_word_list.copy()
      del temp_word_list
      temp_word_list = []

    counter = counter + 1
  # Helps separate word lists & lessens confusion  
  print('______________________________________________________________________')
  
  for word in word_list:
    print(word)

  # Simple statement for understanding the new word bank  
  print('Above are the total possible words: ' + str(len(word_list)) + '\n \n' +
        'Select a word you recognize, because odds are that the unknown ' +
        'words are incorrect \n')
  
  guesses = guesses + 1

else:  
  print('\nOut of guesses ie you suck. Sorry')
```
